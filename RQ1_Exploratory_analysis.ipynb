```python
import pandas as pd
```

```python
# Load data
wave7_csv = "data/preprocessed/filtered_wave_7.csv"
df = pd.read_csv(wave7_csv, low_memory=False)
df
```

```python
# Filter USA data
usa = [840]
df_usa = df[df["Country"].isin(usa)]
df_usa
```

```python
# Display column names line by line
for col in df_usa.columns:
    print(col)
```

```python
# Calculate the age distribution
age_distribution = df_usa["Age"].value_counts().sort_index()

# Display the age distribution
print("Age Distribution:")
print(age_distribution)
```

### Data Cleaning

```python
# Replace missing values (-1, -2, -4, -5) with NaN:
import numpy as np

# Replace missing values with NaN
missing_values = [-1, -2, -4, -5]
df_usa.replace(missing_values, np.nan, inplace=True)
```

```python
# Convert categorical variables into numerical format for analysis
df_usa = pd.get_dummies(df_usa, columns=["Country", "Sex"], drop_first=True)

# Normalize Continuous Variables
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_usa["Age_scaled"] = scaler.fit_transform(df_usa[["Age"]])

# Aggregate confidence scores into an overall trust score
df_usa["Overall_Trust"] = df_usa[["C Government", "C Political parties", "C Courts", "C Elections"]].mean(axis=1)

```

```python
# Check the proportion of missing values for each column:
print(df_usa.isnull().mean().sort_values(ascending=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the proportion of missing values for each column
missing_values = df_usa.isnull().mean().sort_values(ascending=False)

# Plot the missing values as a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x=missing_values.index, y=missing_values.values, palette="Blues_d")
plt.title("Proportion of Missing Values by Column", fontsize=16)
plt.xlabel("Columns", fontsize=14)
plt.ylabel("Proportion of Missing Values", fontsize=14)
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Multivariate Analysis

```python
# Group Analysis: Investigate average confidence in political institutions by Year and Scale of incomes:
df_grouped = df_usa.groupby(["Year", "Scale of incomes"]).mean()
print(df_grouped[["C Government", "C Political parties"]])
```

### Statistical Tests

```python
# Chi-square test for associations between categorical variables:
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df_usa["C Government"], df_usa["C Political parties"])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square test p-value: {p}")
```

```python
# ANOVA for numerical vs categorical relationships:
from scipy.stats import f_oneway

groups = [df_usa[df_usa["C Government"] == i]["Age"].dropna() for i in df_usa["C Government"].unique()]
f_stat, p_val = f_oneway(*groups)
print(f"ANOVA p-value: {p_val}")
```
